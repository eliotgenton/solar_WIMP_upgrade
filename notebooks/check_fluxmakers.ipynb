```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('../paper.mplstyle')

```

```python
oscNext = MCReader('/data/ana/LE/oscNext/pass2/pisa/v02.00/oscNext_pisa_genie_0000_v02.00.hdf5')
charon_flux = np.load('../data/charon_fluxes/ch5-m100_1AU_BRW.npy')
mc_flux = np.load('../data/mc_dn_dz/ch5-m100_oscNext_pisa_genie_0000_v02.npy')
```

```python
ptt  = [12,14,16]
ntt  = ['nu_e', 'nu_mu', 'nu_tau']
nntt = [r'$\nu_{e}$', r'$\nu_{\mu}$', r'$\nu_{\tau}$']
ee = np.logspace(0,2,21)

masks = []
for pt in ptt:
    mask = np.where(oscNext.ptype==pt)[0]
    masks.append(mask)
```

```python
for pt, nt, nnt, mask in zip(ptt, ntt, nntt, masks):
    h0 = np.histogram(oscNext.nu_e[mask], bins=ee, weights=mc_flux[mask])
    h1 = np.histogram(oscNext.nu_e[mask], bins=ee)
    f  = np.divide(h0[0], h1[0])
    plt.step((ee[1:]+ee[:-1])/2, f, where='mid')
    plt.plot(charon_flux['Energy'], charon_flux[nt])
    plt.loglog()
    plt.show()
```
